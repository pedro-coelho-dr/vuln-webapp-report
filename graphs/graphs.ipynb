```python
import pandas as pd
import matplotlib.pyplot as plt

# Ativa dark mode
plt.style.use('dark_background')

# Mapeamento OWASP completo
owasp_labels = {
    'A01:2021': 'A01: Broken Access Control',
    'A02:2021': 'A02: Cryptographic Failures',
    'A04:2021': 'A04: Insecure Design',
    'A05:2021': 'A05: Security Misconfiguration',
    'A07:2021': 'A07: Identification and Authentication Failures'
}

# Lista de vulnerabilidades (atualizada com 10 entradas)
vulns = [
    {'Ttulo': 'Directory Listing', 'Abrangncia': 'client-side', 'OWASP': 'A05:2021'},
    {'Ttulo': 'User Enumeration', 'Abrangncia': 'usurios no autenticados', 'OWASP': 'A07:2021'},
    {'Ttulo': 'Reset sem validao', 'Abrangncia': 'usurios no autenticados', 'OWASP': 'A07:2021'},
    {'Ttulo': 'Tokens previsveis', 'Abrangncia': 'todos os usurios', 'OWASP': 'A02:2021'},
    {'Ttulo': 'Session Fixation', 'Abrangncia': 'usurios no autenticados', 'OWASP': 'A01:2021'},
    {'Ttulo': 'Session Hijacking', 'Abrangncia': 'usurios autenticados', 'OWASP': 'A01:2021'},
    {'Ttulo': 'Verificao de privilgios', 'Abrangncia': 'usurios autenticados', 'OWASP': 'A01:2021'},
    {'Ttulo': 'IDOR', 'Abrangncia': 'usurios autenticados', 'OWASP': 'A01:2021'},
    {'Ttulo': 'Manipulao de preo', 'Abrangncia': 'usurios autenticados', 'OWASP': 'A04:2021'},
    {'Ttulo': 'Falta de MFA / Brute-force', 'Abrangncia': 'todos os usurios', 'OWASP': 'A07:2021'}
]

# Cria DataFrame
df = pd.DataFrame(vulns)

# Substitui cdigo OWASP pelo nome completo
df['OWASP Nome'] = df['OWASP'].map(owasp_labels)

# Paleta customizada
custom_colors = ['#66c2a5', '#fc8d62', '#8da0cb', '#e78ac3', '#a6d854', '#ffd92f']

def plot_pie(col, title):
    counts = df[col].value_counts()
    fig, ax = plt.subplots(figsize=(8, 6))
    wedges, texts, autotexts = ax.pie(
        counts,
        labels=counts.index,
        autopct='%1.1f%%',
        colors=custom_colors[:len(counts)],
        textprops={'color': 'white'}
    )
    ax.set_title(title, color='white', fontsize=14)
    plt.tight_layout()
    plt.show()

# Gerao dos grficos
plot_pie('OWASP Nome', 'Distribuio por Categoria OWASP')
plot_pie('Abrangncia', 'Distribuio por Abrangncia das Vulnerabilidades')
```
