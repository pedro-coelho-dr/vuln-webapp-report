```python
import pandas as pd
import matplotlib.pyplot as plt

# Ativa dark mode
plt.style.use('dark_background')

# Mapeamento OWASP completo
owasp_labels = {
    'A01:2021': 'A01: Broken Access Control',
    'A02:2021': 'A02: Cryptographic Failures',
    'A04:2021': 'A04: Insecure Design',
    'A05:2021': 'A05: Security Misconfiguration',
    'A07:2021': 'A07: Identification and Authentication Failures'
}

# Lista de vulnerabilidades (atualizada com 10 entradas)
vulns = [
    {'Título': 'Directory Listing', 'Abrangência': 'client-side', 'OWASP': 'A05:2021'},
    {'Título': 'User Enumeration', 'Abrangência': 'usuários não autenticados', 'OWASP': 'A07:2021'},
    {'Título': 'Reset sem validação', 'Abrangência': 'usuários não autenticados', 'OWASP': 'A07:2021'},
    {'Título': 'Tokens previsíveis', 'Abrangência': 'todos os usuários', 'OWASP': 'A02:2021'},
    {'Título': 'Session Fixation', 'Abrangência': 'usuários não autenticados', 'OWASP': 'A01:2021'},
    {'Título': 'Session Hijacking', 'Abrangência': 'usuários autenticados', 'OWASP': 'A01:2021'},
    {'Título': 'Verificação de privilégios', 'Abrangência': 'usuários autenticados', 'OWASP': 'A01:2021'},
    {'Título': 'IDOR', 'Abrangência': 'usuários autenticados', 'OWASP': 'A01:2021'},
    {'Título': 'Manipulação de preço', 'Abrangência': 'usuários autenticados', 'OWASP': 'A04:2021'},
    {'Título': 'Falta de MFA / Brute-force', 'Abrangência': 'todos os usuários', 'OWASP': 'A07:2021'}
]

# Cria DataFrame
df = pd.DataFrame(vulns)

# Substitui código OWASP pelo nome completo
df['OWASP Nome'] = df['OWASP'].map(owasp_labels)

# Paleta customizada
custom_colors = ['#66c2a5', '#fc8d62', '#8da0cb', '#e78ac3', '#a6d854', '#ffd92f']

def plot_pie(col, title):
    counts = df[col].value_counts()
    fig, ax = plt.subplots(figsize=(8, 6))
    wedges, texts, autotexts = ax.pie(
        counts,
        labels=counts.index,
        autopct='%1.1f%%',
        colors=custom_colors[:len(counts)],
        textprops={'color': 'white'}
    )
    ax.set_title(title, color='white', fontsize=14)
    plt.tight_layout()
    plt.show()

# Geração dos gráficos
plot_pie('OWASP Nome', 'Distribuição por Categoria OWASP')
plot_pie('Abrangência', 'Distribuição por Abrangência das Vulnerabilidades')
```
